### 2. 목적성
#### 사과 1개 가격이 1,000원인데 사과 개수가 0개, 1개, 2개, ... , N개로 매번 다르게 불러져서 총 사과 가격을 알고 싶다고 해보자

```python
def apple_cal(apple_num):
    total_price = 1000 * apple_num
    return total_price
```

#### 사과를 사는 손님이 23개의 사과를 구매했을 때 총 가격은?

```python
num = 23
total_price = apple_cal(apple_num = num)
print("손님이 구매한 사과 개수는 ", num, "개 이고 총 가격은 ", total_price, "원 입니다.")
```

#### 만약 1500개의 사과를 구매한다면?

```python
num = 1500
total_price = apple_cal(apple_num = num)
print("손님이 구매한 사과 개수는 ", num, "개 이고 총 가격은 ", total_price, "원 입니다.")
```

#### 즉, 입력받는 값이 매번 변할 수 있는데 그에 맞는 계산 목적을 가지는 경우 변수가 필요함
#### 나중에 인공지능, 빅데이터 분석 등을 하는 경우에 매번 필요할 정도로 중요함

### 변수 이름 규칙 1. 영문과 한글 및 숫자만 사용할 수 있음

```python
a = 100
abcd = 200
zzzz = 300
p1234 = 400
its4you = 600
고영민 = 2500

print(a)
print(abcd)
print(zzzz)
print(p1234)
print(its4you)
print(고영민)
```

#### 단, 영문과 한글로 시작해야함, 숫자로 시작하는 경우 혹은 숫자만으로 이루어진 변수는 오류가 발생함

```python
3333 = 100
```

#### 변수명이 숫자로 시작하는 경우도 오류가 발생함

```python
3abcd = 200
```
