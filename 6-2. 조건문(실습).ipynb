# 1. if 문
#### if문은 가장 간단한 형태의 조건문으로써 
#### - 조건식이 참일때는 참인 조건식에 대응하는 문장을 실행하고
#### - 조건식이 거짓일때는 아무것도 실행하지 않음

```python
num = 99
if num < 100 :
    print("100보다 작습니다.")
```

#### 
#### * if 조건식이 참일 때 수행되는 문장은 여러 개가 올 수 있으며 반드시 들여쓰기된 상태에서 작성해야 함
#### * 들여쓰기 칸 수도 같아야함

```python
num = 200

if num > 100 :
    print("200은")
    print("100보다")
    print("큽니다.")
```

#### 들여쓰기가 잘못된 경우 -> if 조건식 : 밑에는 항상 들여쓰기가 되어야 오류가 발생하지 않음

```python
number = 100
if number == 10:
print("10입니다.")
```

#### 다시 올바르게 표현하면

```python
number = 100
if number == 10:
    print("10입니다.")
```

# 2. if ~ else 문
#### if ~ else 문은 조건식의 참과 거짓에 대해 대응하는 문장을 실행하게 하기 위한 조건문임
#### - 조건식이 참일때는 참인 조건식에 대응하는 문장을 실행하고
#### - 조건식이 거짓일때는 거짓인 조건식에 대응하는 문장을 실행함


```python
num = 200

if num < 100 :
    print("100보다 작군요.")
else :
    print("100보다 크군요.")
```

#### 
#### 

#### * 마찬가지로 if와 else 각각에 대응하는 실행 문장들은 여러 개가 올 수 있으며,
#### * 들여쓰기가 같아야함

```python
num = 200

if num < 100 :
    print("100보다 작군요.")
    print("여기는 참일 때 실행되는 문장입니다.")

else :
    print("100보다 크군요.")
    print("여기는 거짓일 때 실행되는 문장입니다.")
    
print("프로그램 종료")
```

### - 중첩 if ~ else 문
#### 앞서 배운 if 문과 , if ~ else 문으로는 많아야 참, 거짓인 두 가지 경우밖에 제시하지 못했음
#### 
#### 예를 들어, 전주에 사는 학생 중에서 25세 이상인 학생이 몇 명인지 알고 싶다고 해보자.
#### 그렇다면 질문은 

#### 이와 같이 조건문을 중첩으로 사용하여 여러 가지의 경우를 나타낼 수 있음

#### 
#### 예를 들어 입력한 숫자가
#### 1. 0 이상 100과 같거나 작으면 "0 과 100 사이"
#### 2. 0 미만이면 "0 보다 작음"
#### 3. 100 보다 크면 "100보다 큼"

#### 을 중첩 조건문을 사용하면 다음과 같이 표현할 수 있음

```python
x = int(input("숫자 입력 :"))

if (x >= 0) and (x <= 100) :
    print("0과 100 사이")
else :
    if x < 0 :
        print("0 보다 작음")
    else :
        print("100보다 큼")
```

# 3. if ~ elif ~ else 문
#### 아래와 같이 시험성적에 대해 등급을 나타내는 조건문을 if ~ else를 중첩을 통해 표현할 수 있음


```python
score = int(input("시험성적 : "))

if (score >= 80) and (score <= 100) :
    print("수")
else :
    if (score >= 60) and (score < 80) :
        print("우")
    else :
        if (score >= 40) and (score < 60):
            print("미")
        else :
            if (score >= 20) and (score < 40):
                print("양")
            else :
                print("가")
```

#### 하지만 이와 같이 중첩된 if문을 반복 사용하면 프로그래밍 코드가 복잡해보이고
#### 코드 오류가 일어날 확률이 높아짐
#### 이렇게 중첩된 조건문의 복잡성을 줄이기 위해 if ~ elif ~ else 문이 등장함


#### 
#### - 문자열을 조건문으로 받는 경우도 가능함

```python
score = int(input("시험성적 : "))

if (score >= 80) and (score <= 100):
    print("수")
elif (score >= 60) and (score < 80):
    print("우")
elif (score >= 40) and (score < 60):
    print("미")
elif (score >= 20) and (score < 40):
    print("양")
else:
    print("가")
```
