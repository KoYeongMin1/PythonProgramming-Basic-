#### - 루트($\sqrt{}$) 제곱근 연산 : math 라이브러리를 사용한 sqrt() 함수 사용법
#### math 라이브러리는 다양한 수학 기호들을 프로그래밍하여 만들어둔 라이브러리로써 sqrt() 제곱근 등을 사용할 수 있는 라이브러리임
#### $\sqrt{a^{2} + b^{2}} = \sqrt{c^{2}}$

```python
import math

a = 3
b = 4
c2 = a**2 + b**2

c = math.sqrt(c2)
print(c)
```

#### 혹은 라이브러리를 사용하지 않고 제곱 ** 만을 이용하여 계산하는 것도 가능함
#### $\sqrt{c} = c^{\frac{1}{2}} = c^{0.5}$

```python
a = 3
b = 4
c2 = a**2 + b**2

c = c2**(1/2)
print(c)
```

### - 명제 정의 :
#### 논리학적으로 뜻이 분명한 문장을 말한다. 즉, '참' 혹은 '거짓' 임을 검증할 수 있는 객관적 사태가 포함된 문장을 말한다.
# 

### - 비교연산자의 정의 :
#### 피연산자인 수를 대상으로 상대적인 크기를 판단하여 참(True) 과 거짓(False)을 반환함

### - 비교연산자에서의 연산자와 피연산자 :
#### 연산자 : >, <, =, <=, >= ...
#### 피연산자 : 수
#### 결과값 : 명제(불) True, False

# 
### - 프로그래밍에서의 비교연산자 소개

#### * 같다 (==) 와 같지 않다(!=)
#### * <=, >= 순서 : 위가 왼쪽, 아래가 오른쪽


#### 
#### 피연산자 수 100과 200을 대상으로 비교연산자를 사용한 결과 : 그 결과값이 명제인 (True or False) 로 나옴

```python
print(100 < 200)
```

```python
print(100 == 200)
```

#### 피연자사 100과 200이 비교연산자 != 을 만나 100과 200이 다른가요? 라는 의미이며 다르므로 결과값이 참인 True 가 나옴

```python
print(100 != 200)
```

#### 같거나 크다 >= , 같거나 작다 <= 는 프로그래밍에서 대소를 비교하는 >,<가 먼저 나오고 뒤에 = 이 붙음

```python
a = 50
b = 50
print(a >= 50)
```

### - 논리연산자의 정의 :
#### 명제를 피연산자로 논리합(OR), 논리곱(AND) , 부정(NOT) 을 연산자로 하여 결과값이 명제인 연산자

### - 집합연산자의 정의 :
#### 피연산자를 집합으로 하고 연산자를 합집합($\cup$), 교집합($\cap$), 차집합(C) 등을 사용하여 그 결과값이 집합인 연산자
### - 집합 :
#### 특정한 조건에 맞는 별개의 원소들의 모임(명확한 기준에 의하여 주어진 서로 다른 대상들이 모여 이루는 새로운 대상)
#### 파이썬에서는 집합을 하나의 데이터형인 'set' 으로 정의해놓음

### - 집합연산자에서의 연산자와 피연산자 :
#### 연산자 : $\cup, \cap$ ,...
#### 피연산자 : 집합
#### 결과값 : 집합

#### 
#### 데이터형에서 집합연산자를 사용하기 위한 데이터형 set


#### 집합 연산자 기호와 파이썬 명령어


### - 집합(set) 정의 : {} 
#### 파이썬에서 집합은 { } 안에 원소를 정의하여 사용할 수 있음

```python
A = {1, 2, 3}

print(A)
print(type(A))
```

#### 문자열도 집합의 원소로 사용이 가능함

```python
A = {'a', 'b', 'c'}

print(A)
print(type(A))
```

#### * 집합 set 은 중복요소를 제거함

```python
A = {'a' ,'a', 'b', 'c'}

print(A)
print(type(A))
```

#### * 집합 set은 별개의 원소이기만 하면 모두 정의 가능

```python
A = {'a', 3, 'c', 6}

print(A)
print(type(A))
```

#### 
### - 집합연산자 예시

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a)
print(b)
```

#### a 집합과 b 집합의 합집합($\cup$)은 키보드에서 쉬프트 백슬래시 을 사용한 기호를 사용함

```python
print(a | b)
```

#### a 집합과 b 집합의 교집합($\cap$)은 & 기호를 사용함

```python
print(a & b)
```

#### a 집합과 b 집합의 차집합은 - 기호를 사용함

```python
print(a - b)
```

#### /= 복합연산자

```python
num = 6
num /= 2
print(num)
```

/= 복합연산자는 변수 num을 어떤 수로 나눈 몫의 값을 다시 num 변수에 저장하는 연산자임

```python
num = 6
num = num / 2
print(num)
```

#### 
#### //= 복합연산자 
#### // 은 몫을 구하는 / 연산자와 같은 역할을 하지만 소숫점을 버리고 정수형으로 반환하는 연산자임
7에 대하여 2로 나눈 몫을 / 와 // 연산자로 구한 결과 : // 의 경우 몫을 정수형으로 바꾸면서 소숫점 5를 없앤 것을 알 수 있음

```python
num = 7
print(7 / 2)

print(7 // 2)
```

```python
num = 6
num //= 2
print(num)
```

##  * 연산자의 우선순위
#### 우리가 이상형을 고른다고 가정해보자
#### 예를 들어, 우선 외모가 이뻐야하고 그 다음으로는 마음이 착해야 하고 그 다음은 부자여야하고, ... 등등
#### 이렇게 조건들 중에서도 우선적으로 먼저 만족해야하는 조건들이 있음
#### 파이썬에서 연산자도 그 우선순위의 순서가 정해져 있으며
#### 중요도에 따라 바꿀 수도 있음

파이썬에서 비트 연산자의 사용 이유:
1. 비트 연산자는 연산 속도가 빠릅니다.   
비트 연산자는 비트 단위로 연산을 수행하기 때문에, 더 작은 연산 단위에서 연산을 처리할 수 있습니다. 이에 따라 비트 연산자는 일반적인 산술 연산자보다 연산 속도가 빠르고, 대용량 데이터 처리에 적합합니다.


2. 비트 연산자는 이진수 연산에 적합합니다.   
비트 연산자는 이진수 연산에 적합합니다. 이진수 연산은 컴퓨터 하드웨어에서 많이 사용되는 연산 방법으로, 데이터를 0과 1로 표현하는 방식에 맞게 처리할 수 있습니다.


3. 비트 연산자는 데이터 압축, 암호화 등에 사용됩니다.   
비트 연산자는 데이터 압축, 암호화 등에 활용될 수 있습니다. 예를 들어, 데이터를 이진수로 표현한 후 비트 연산을 수행하여 데이터를 압축하거나, 암호화 알고리즘에서 비트 연산을 활용하여 데이터를 보호할 수 있습니다.

### ~ 기호 : 비트NOT 연산자
#### 비트연산자이며, 이를 설명하기 위해 10진수 표현 3을 2진수로 표현했을 때
#### bin() 함수를 사용하여 10진수 표현을 bin() 을 통해 2진수 표현으로 변환할 수 있음
#### 앞쪽에 0b 는 2진수 시작을 알리는 파이썬 관례임

```python
print(bin(10))
```

#### 2진수 표현 3을 ~ 비트 NOT 연산자를 사용한 결과

```python
print(bin(~10))
```
