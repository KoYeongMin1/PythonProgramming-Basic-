### 클래스가 필요한 이유를 계산기 프로그래밍으로 알아보자
- 계산기는 이전에 계산한 결괏값을 기억하고 있어야함.
- 함수(def)를 사용하여 계산기의 '더하기' 기능을 구현할 수 있음


- 계산기가 이전의 결과값을 저장하기 위해 global 전역변수를 선언하여 다룰 수 있음

```python
result = 0

def add(num):
    global result
    result += num
    return result
```

```python
print(add(3))
print(add(4))
```

- 만약 2개의 다른 계산기를 동시에 사용하여 각각이 이전에 계산된 결과값을 저장하고 있으려면 어떻게 해야할까?

```python
# calculator2.py
result1 = 0
result2 = 0

def add1(num):  # 계산기1
    global result1
    result1 += num
    return result1

def add2(num):  # 계산기2
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

- 하지만 계산기가 2개 이상 더욱 많아지는 경우에는 어떻게 구현해야 효율적으로 구현이 가능할까?

- class를 사용하여 구현한 계산기 예시

```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

- 위와 같이 여러 개의 global 전역변수를 따로 선언하지 않고도 사용할 수 있는 효율성을 가지고 있음을 볼 수 있음

### 클래스로 계산기 만들어보기


#### 클래스로 사칙연산 계산기를 구현해보자

```python
class Cal:
    def __init__(self, x):
        self.x = x
    
    def add(self, y):
        self.x = self.x + y
        return self.x
    
    def sub(self, y):
        self.x =  self.x - y
        return self.x
    
    def mul(self, y):
        self.x =  self.x * y
        return self.x
    
    def div(self, y):
        self.x = self.x / y
        return self.x
```

#### 객체 생성 : 초깃값은 0

```python
cal = Cal(x=0)
```

#### 초기값 0에 5를 더한 경우

```python
cal.add(y=5)
print(cal.x)
```

#### 더해진 5에 2를 빼는 경우

```python
cal.sub(y=2)
print(cal.x)
```

#### 계산된 3에 4를 곱하는 경우

```python
cal.mul(y=4)
print(cal.x)
```

#### 계산된 12에 3으로 나눈 경우

```python
cal.div(y=3)
print(cal.x)
```

#### 모든 계산이 완료되면 다시 초깃값 0으로 초기화

```python
cal.x = 0
print(cal.x)
```

```python
print(type(cal))
```

### 클래스에서는 상위 개념의 클래스를 상속받아 그 속성과 함수(메소드)를 그대로 사용할 수 있음

#### 예를 들어 바퀴 수와 움직이는 행동을 하는 운송수단 클래스가 있다고 해보자

```python
class Vehicle:
    def __init__(self, tire):
        self.tire = tire
        
    def move(self):
        print("굴러감")
```

```python
vehicle = Vehicle(tire=4)
```

```python
print(vehicle.tire)
```

```python
vehicle.move()
```

#### - 행동으로 굴러감을 가진다고해보자

```python
martiz = Martiz()
```

## 
### 클래스 상속을 사용하여 운송수단을 마티즈에 넣어주기
#### class 클래스이름(상속받을 클래스 이름): -> 으로 간단히 상속 받을 수 있음

```python
class Vehicle:
    def __init__(self, tire):
        self.tire = tire
        
    def move(self):
        print("굴러감")

class Martiz(Vehicle):
    def __init__(self, mc="대우자동차", tire=4):
        super().__init__(tire)
        self.mc = mc

martiz = Martiz()
```

```python
martiz.mc
```

```python
martiz.tire
```

```python
martiz.move()
```
