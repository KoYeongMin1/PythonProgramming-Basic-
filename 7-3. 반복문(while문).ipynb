#### 전주대학교 건물들은 24시간 동안 에스원 경비가 돌아가고 있음.
#### 
#### 만약 이를 유한한 반복문으로 구현해본다면

```python
import time

for i in range(5):
    print("경비 가동 중")
    time.sleep(1)
```

- time 이라는 라이브러리는 python에서 시간을 다루는 함수들을 모아놓은 라이브러리임.   
- time 라이브러이에 sleep() 이라는 함수는 프로그램 명령이 잠깐 정지 혹은 잠들었다가 깨어나서 수행하라는 함수임.   
- sleep() 안에 받는 값은 초로 인지하여 sleep(1) 인 경우 1초 동안 쉬고 명령을 수행하라는 의미임.

#### 더 오래 에스원 경비가 돌아가게끔 만들고 싶다면 range() 값을 키우면됨.

```python
import time

for i in range(100000000000):
    print("경비 가동 중")
    time.sleep(1)
```

## 반복문 while문

#### 유한한 반복을 사용하는 for문과 달리 while 문은 무한 반복을 할 수 있음

```python
import time

while(True):
    print("경비 가동 중")
    time.sleep(1)
```


- while(조건식) 에서 while 은 "~ 하는 동안" 이고, 조건식이 참(True) 이면 계속 반복한다는 의미임
- 위와 같이 조건식 자리에 True 자체를 집어 넣으므로써 계속 True 이므로 무한 반복이 실행됨
#### 무한하게 반복된다면 무한한 반복을 멈추거나 종료하는 방법 또한 필요함
1. jupyter notebook 에서 사각형(interrupt the kernel) 을 눌러 멈추는 방법
2. 키보드의 특정 키를 입력받아 종료하는 방법
3. 카운트 변수를 추가하여 종료하는 방법

#### 
### 1. interrupt the kernel 방법

```python
import time

while(True):
    print("멈춰줘")
    time.sleep(0.5)
```

#### 
### 2. 키보드의 특정 키를 입력받아 종료하는 방법
- keyboard 라이브러리는 키보드를 입력받아 처리하는 라이브러리임.
- keyboard 라이브러리 안에 is_pressed() 라는 함수는 어떤 키를 입력받았을 때 명령을 수행하는 함수임.
#### 아래는 q 를 누르는 경우 반복문을 break 종료하는 코드

```python
import keyboard

while True:
    if keyboard.is_pressed('q'):
        print("q를 입력받아서 반복문을 종료합니다.")
        break
```

- anaconda prompt 를 사용하여 keyboard 라이브러리 설치하는 명령어로 pip install keyboard 를 수행

```python
import keyboard

while True:
    if keyboard.is_pressed('q'):
        print("q를 입력받아서 반복문을 종료합니다.")
        break
```

- is_pressed() 함수 안에는 키보드로 입력받을 수 있는 모든 input이 가능함.
#### k를 입력받아 종료되는 반복문 예시

```python
import keyboard

while True:
    if keyboard.is_pressed('k'):
        print("종료")
        break
```

#### 
### 3. 카운트 변수를 도입하여 종료하는 방법
#### 무한 반복이 수행되는 동안 카운트 변수를 증가시켜 특정값을 만족할 경우 종료하는 반복문을 작성해볼 수 있음

```python
c = 0 # 카운트 변수 도입

while(True):
    if c == 5 : 
        break
    print("멈춰줘")
    c = c + 1
    print(c)
```

#### 
#### 카운트 변수를 사용하여 0부터 10까지 출력하고 종료하는 while 문
- break가 어디에 위치하는가에 따라 결과가 달라지므로 주의

```python
c = 0

while(True):
    print(c)
    c = c + 1
    if c == 11 :
        break
```

```python
c = 0

while(True):
    print(c)
    if c == 11 :
        break
    c = c + 1
```

## 조건식을 활용한 while 문

### 조건식을 활용한 while 문의 경우, 카운트 변수가 필요함
#### 아래와 같이 카운트 변수 c 가 10보다 작은 경우에만 반복하는 while문은 다음과 같음

```python
c = 0

while(c < 10):
    print(c)
    c = c + 1
```

#### * 카운트 변수는 꼭 0을 대입할 필요 없음
#### * 카운트 변수를 도입하였지만 반복문 내에 카운트 변수를 변화시키지 않는다면 종료되지 않음에 주의

#### 
#### 문제 : while문을 사용하여 2부터 10까지 출력해보세요

#### 
#### while 문을 사용하여 1부터 100까지 모두 더한 값을 나타낼 수 있는데 이때는 변수가 2개가 필요함
#### 카운트 변수와, 모두 더한 값을 저장할 값

```python
c = 1 # 카운트 변수
total = 0 # 모두 더할 값

while(c <= 100):
    total = total + c # 누적된 total 에 c를 더한 값
    c = c + 1 # c 를 하나씩 늘려나감
print("1부터 100까지 모두 더한 값 : ", total)
```

####  문제 : 0부터 100까지 짝수의 합을 모두 더한 값을 출력해보세요
