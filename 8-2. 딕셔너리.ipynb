```python
score_dict[''] = 'A'
print(score_dict)
```

####

###

```python
for key in score_dict:
    print(key)
```

```python
for key in score_dict:
    print(score_dict[key])
```

```python
for key in score_dict:
    if score_dict[key] == 'C':
        score_dict[key] = 'A'
    
    print(score_dict[key])
```
