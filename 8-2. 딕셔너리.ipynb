### 사전 사용 방법

## 
## 프로그래밍에서 딕셔너리
- 프로그래밍에서 딕셔너리는 이렇게, 유니크하게 구분할 수 있는 키(Key) 와
- 키(Key) 에 대응하는 값(Value) 을 한 쌍으로,
- { } 안에 여러 개의 키와 값의 쌍을 가지고 있는 데이터형임
#### 
#### 백과사전과 프로그래밍 딕셔너리 비교


## 
### 딕셔너리 선언
- 중괄호 { } 로 선언하며,
- { } 안에 Key 값과 Value 값을 콜론 ":" 으로 구분하여 한 쌍을 만들어줘야함 
- { Key : Value }

```python
word_dict = { "사과" : "사과나무의 열매" }
print(word_dict)
```

#### 
- 딕셔너리는 여러 개의 key : value를 넣을 수 있고 쉼표로 구분하며 항상 쌍을 이뤄야함

```python
word_dict = { "사과" : "사과나무의 열매", "배" : "배나무의 열매"}
print(word_dict)
```

#### 
### 딕셔너리 Key 값과 Value 값 보는 방법
- 딕셔너리 클래스 내부에는 Key 값과 Value 값을 보는 함수가 있음

#### 1. 딕셔너리 Key 값을 보려면
- 딕셔너리.keys() 함수 사용

```python
word_dict.keys()
```

#### 2. 딕셔너리 Value 값을 보려면
- 딕셔너리.values() 함수 사용

```python
word_dict.values()
```

#### 3. 딕셔너리의 Key 와 Value 모두 보려면
- 딕셔너리.items() 함수 사용

```python
word_dict.items()
```

## 
### 딕셔너리의 Key 와 Value 데이터형
- Key 와 Value 값은 int, float, str 등 다양한 데이터형을 사용할 수 있음
- 숫자(int, float)과 문자(str) 을 자유롭게 사용 가능함

#### 딕셔너리 Key, Value 데이터형 조합


#### 1. Key 와 Value 모두 숫자인 경우
예 : 1등은 100만원, 2등은 50만원 3등은 30만원의 상금을 주려고 하는 경우

```python
A = {1:100, 2:50, 3:30}
print(A)
```

#### 
#### 2. Key는 숫자, Value는 문자인 경우
예 : 1등은 고영민, 2등은 김광배, 3등은 최문열로 등록할 경우

```python
A = {1 : '고영민', 2 : '김광배', 3 : '최문열'}
print(A)
```

#### 
#### 3. Key는 문자, Value는 숫자인 경우
예 : 학생이름과 성적

```python
A = {'고영민' : 50, '김광배' : 100, '최문열' : 80}
print(A)
```

#### 4. Key와 Value 모두 문자인 경우

```python
word_dict = { "사과" : "사과나무의 열매", "배" : "배나무의 열매"}
print(word_dict)
```

### 딕셔너리 내에 데이터 접근 방법
- 딕셔너리는 리스트, 튜플과 다르게 순서 기반인 인덱스가 아니라 키(Key) 값을 사용하여 데이터에 접근함
- 딕셔너리[Key값] 으로 Key 값에 대응하는 Value 값에 접근할 수 있음

```python
A = {'a' : 100, 'b' : 80, 'c':50, 'd':90}
print(A)
```

예를 들어, 위의 딕셔너리에서 d 의 성적이 궁금하다면?

```python
print(A['d'])
```

즉 딕셔너리[ ] 괄호 안에 Key 값을 넣어 접근해볼 수 있음. a,b,c,d 모두 값을 출력하면 다음과 같음

```python
print(A['a'])
print(A['b'])
print(A['c'])
print(A['d'])
```

#### 문제 : 다음과 같은 일본언어문화학과 학생들의 학점 딕셔너리가 있다고 해보자 

```python
score_dict = {'국유니' : 'A',
             '정보승' : 'B',
             '이송현' : 'A',
             '신은지' : 'C',
             '방수정' : 'C',
             '김승구' : 'A',
             '국유니' : 'A',
             '김예훈' : 'B',
             '박세종' : 'C', 
             '고영민' : 'D',
             '수상한A' : 'B',
             '수상한B' : 'C',
             '수상한C' : 'D'}
print(score_dict)
```

#### 또한 딕셔너리에서는 Value 값을 바꿀 수 있음
- 딕셔너리[key값] = 새로운 값

고영민의 학점을 A 로 변경한다면 다음과 같음

```python
score_dict['고영민'] = 'A'
print(score_dict)
```

#### 
#### 문제 : 학점이 C인 학생들의 성적을 모두 A로 바꿔서 출력해주세요

#### 
### 딕셔너리와 반복문을 함께 사용하는 방법
- 반복문 요소에 딕셔너리를 넣게될 경우 출력은 Key 값들만 출력하게 됨

```python
for key in score_dict:
    print(key)
```

#### Value 값들을 보려면 이 key 값들을 다음과 같이 사용할 수 있음

```python
for key in score_dict:
    print(score_dict[key])
```

그럼, 학점인 C 인 친구들을 A 로 바꿔주는 방법을 반복문을 사용한다면 다음과 같이 할 수 있음

```python
for key in score_dict:
    if score_dict[key] == 'C':
        score_dict[key] = 'A'
    
    print(score_dict[key])
```

이제 학생 이름과 학점을 모두 출력해보자

```python
print(score_dict)
```
