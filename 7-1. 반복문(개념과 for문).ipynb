## 유한한 반복 : for 문 개념과 사용법

반복에 사용될 유한한 요소들이 [1,2,3] 세 개의 숫자가 있다고 해보자

```python
for i in [1,2,3]:
    print(i)
```

### for문의 진행순서 : 


#### * 즉 요소들을 순차적으로 변수 i 에 넣어 반복함
#### * 요소 [1,2,3] 을 i 에 in(안으로) 대입해라 의미로 해석

#### 
#### 문제 : 1부터 5까지 순차적으로 출력하는 반복문을 만들어보세요

#### 
### for문에서 반복될 때, 변수 i 뿐만 아니라 추가적으로 다른 문장들을 수행 가능함

```python
for i in [1,2,3]:
    print(i)
    print("반복됨")
```

### for 문의 진행순서: 


#### 
#### 1등부터 3등까지 "참 잘했어요" 를 출력하고자 한다면

```python
for i in [1,2,3]:
    print(i, "등")
    print("참 잘했어요!")
```

#### 
#### * 추후에 배울 리스트 [] , 튜플 (), 집합 {} 등 여러 개의 요소들을 가지는 클래스라면 for 문에 모두 사용 가능함
#### 리스트 list : [ ] 예시

```python
for i in [1,2,3]:
    print(i)
```

#### 튜플 tuple : ( ) 예시

```python
for i in (1,2,3):
    print(i)
```

#### 집합 set : { } 예시

```python
for i in {1,2,3}:
    print(i)
```

### 
#### * for 문의 변수 i 는 다른 변수로 지정하여 사용 가능함
#### 아래와 같이 변수 이름 규칙에 맞는 변수들은 모두 활용 가능함

```python
for j in [1,2,3]:
    print(j)
```

```python
for food in ['햄버거', '비빔밥', '삼겹살']:
    print(food)
```

```python
for 나이 in [15, 19, 44]:
    print(나이)
```

#### 초등학교 반 학생들의 키가 다음과 같을 때 탈 수 있는 사람과 탈 수 없는 사람을 구분해보자


```python
children = [151, 125, 146, 137, 143, 139]

for i in children:
    if i < 140 :
        print(i, "cm 이므로 탑승 불가")
    else :
        print(i, "cm 이므로 탑승 가능")
```

### - range() 함수 소개
#### - 만약 100까지 출력하라고 한다면, 리스트 [ ] 에 들어가는 요소를 1부터 100까지 입력해야할 것임
#### - 우리는 1부터 한 단위씩 증가하여 100까지 라는 규칙을 알고 있으므로 이를 표현하는 기능이 있으면 좋을 것임
#### - 이런 기능을 하는 range() 함수가 있음

```python
for i in range(1,11):
    print(i)
```

#### 
### range() 함수 : 
#### 파이썬에서 range() 함수는 반복문과 함께 많이 쓰이는 함수로써, 지정된 범위의 값을 정해진 규칙대로 출력하는 함수임
#### 
#### range() 를 print() 로 출력해본 경우 : 단순히 print() 출력을 하였을 때는 전체가 보이지 않음

```python
print( range(20) )
```

#### range() 의 데이터형을 출력해본 경우 : range() 의 데이터형은 range

```python
print( type(range(20)))
```

### range() 함수 3가지 종류
#### - range() 함수는 3가지 방법으로 인자를 받을 수 있으며, 모든 인자는 데이터형이 정수(int) 여야함

#### range( a ) : 0부터 a-1 까지 나타냄

```python
a = 3
for i in range(a):
    print(i)
```

#### range( a, b ) : a부터 b-1 까지 나타냄

```python
a = 3
b = 5

for i in range(a,b):
    print(i)
```

#### range( a, b, c ) : a부터 b-1 까지 c 간격으로 나타냄

```python
a = 2
b = 7
c = 2

for i in range(a,b,c):
    print(i)
```

### 새로운 변수를 도입하여 반복문을 응용하는 경우

#### 1부터 10까지 총 합을 구하는 경우를 생각해보자

```python
for i in range(1,11):
    print(i)
```

#### 위와 같이 1부터 10까지 출력은 되었으나 1부터 10까지 순차적으로 누적하여 총합을 구하려면 새로운 변수를 도입할 수 있음

```python
total = 0

for i in range(1,11):
    total = total + i
print("1부터 10까지 총 합 :" ,total)
```

#### total = 0 라는 새로운 변수를 도입하여,
#### i 에 순차적으로 대입되는 값을 누적시킬 수 있음

#### 
#### 팩토리얼 함수를 for문을 사용하여 구현할 수 있음


```python
x = input("수를 입력 :")
x = int(x)

total = 1
for i in range(1, x+1):
    total = total * i
print(x, "!  = ", total)
```

### 중첩 반복문

#### - for 반복문을 2개 이상 함께 사용할 수 있음
#### - 중첩 반복문의 순서는 다음과 같음

```python
for i in [1,2]:
    for j in [3,4]:
        print(i, j)
```

#### * 상위 반복문부터 시작하지만, 하위 반복문을 전부 반복한 뒤, 상위 반복문의 다음 반복이 이뤄짐

#### 
#### for문을 사용하여 구구단(2단-9단)을 만들어보자


### 
#### 사과, 바나나, 배의 가격을 각각 1,2,3개에 대해 출력하는 것을 중첩 반복문으로 작성하면


```python
for i in [1000, 1500, 2000]:
    for j in [1,2,3]:
        print("가격 : ", i, "원, 개수 : ", j, " 개, 총 가격 : ", i*j , "원")
```
