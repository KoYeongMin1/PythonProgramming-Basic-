## 데이터형이란?
### - 데이터형, 데이터 타입(data type) 또는 자료형은 프로그래밍 언어에서 여러 종류의 데이터를 식별하는 분류다.
### - 더 나아가 해당 데이터형에 대한 가능한 값, 해당 데이터형에서 수행을 하는 명령들, 해당 데이터형을 저장하는 방식을 결정한다.
## 

## 

### 파이썬에서 데이터형은 몇 가지로 구분하고 약속해놓았음


### 데이터형 종류와 특징
#### - 기본 데이터형(정수,실수, 문자열) : 가장 기본적인 데이터 형태를 나타냄
#### - 리스트, 튜플, 딕셔너리 : 빅데이터를 다루기 용이
#### - 변수 : 프로그래밍에서 변수는 임의의 값을 저장한 메모리 주소임
#### - 클래스 : 필요에 따라 새로운 데이터형을 만들어내기 위함

# 데이터형과 데이터형을 다루는 방법을 통해 만들 수 있는 프로그램들 예시

## - 음료수 자판기

# 
### 예를 들어, 파이썬으로 음료수 자판기를 만들어보면

```python
def drink_machine(drink_name):
    drink_list = {'포카리스웨트' : 1200 , '웰치스' : 1100 , '멕스웰' : 800, '코코팜' : 1000}
    if drink_name in drink_list :
        print("선택하신 음료 {} 는 {} 원입니다.".format(drink_name, drink_list[drink_name]))
    else :
        print("선택하신 음료는 없습니다.")
```

```python
drink_name = input("음료수를 선택해주세요 (포카리스웨트, 웰치스, 멕스웰, 코코팜) :\n\n")
drink_machine(drink_name=drink_name)
```

## - 대학교 학점 관리 프로그램


```python
c = 0 
stu_score = {}

print("=================================================")
print("=================학점관리프로그램================")
print("=================================================")

while(1):
    print('\n')
    command = input("종료 혹은 학생이름을 입력하세요 :")
    print('\n')
    if command == '종료' :
        if c == 0 :
            print("종료")
            break
        else :
            print("\n========평균 학점 성적순 나열========\n")
            print(sorted(stu_score.items(), key=lambda x:x[1], reverse=True))
            break
    
    else :
        A_s = float(input("A 과목 학점을 입력하세요 : "))
        B_s = float(input("B 과목 학점을 입력하세요 : "))
        C_s = float(input("C 과목 학점을 입력하세요 : "))
        
        avg = (A_s + B_s + C_s) / 3
        stu_score[command] = round(avg,2)
        
        c = c+1
```
