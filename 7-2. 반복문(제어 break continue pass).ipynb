#### 
- 다음과 같은 상황에 사람이 들어갈 경우 자칫 생명을 잃을 수 있는 위험한 상황이 야기될 수 있음
- 기계가 반복 작업을 하다 위험 신호가 잡히면 바로 멈춰야 함
- 하지만 현재 배운 반복문을 사용한 방법으로는 어려움
- 아래와 같이 공장의 위험 신호를 받는 반복문이 있다고 해보자

```python
import time

factory_sign = ['정상', '정상', '정상', '위험', '정상', '정상']

for i in factory_sign:
    print(i)
    time.sleep(1)
```

#### 위의 import time은 시간을 관리하는 코드를 모아놓은 라이브러리로써, time.sleep(1) 은 코드가 실행될 때 1초 동안 멈추고 진행하라는 의미임 

#### 
- 위와 같은 경우에는 이미 위험 신호를 받았음에도 불구하고 기계가 멈추지 않아 사람이 죽은 상황임
- 우리는 이와 같이 반복문이 수행되다가 어떤 특정한 상황에 마주하게 되면 반복문을 멈추는 상황이 필요함
-  break 제어 명령어를 통해 반복문을 중단하는 예시는 아래와 같음

```python
import time

factory_sign = ['정상', '정상', '정상', '위험', '정상', '정상']

for i in factory_sign:
    if i == '위험':
        print("위험 감지! 기계를 정지합니다.")
        break
    print(i)
    time.sleep(1)
```

## 1. break 제어 명령어

```python
for i in range(10):
    if i == 4:
        break
    print(i)
```

### 반복문 안에서 break를 만나게될 경우, 반복문이 끝나지 않았음에도 불구하고, break(멈춤) 으로 인해 반복문이 종료됨

## 2. continue 제어 명령어

#### 여러분이 공장의 로봇을 프로그래밍하는 프로그래머가 되었다고 해보자.
#### 여러분이 취직한 공장은 다음과 같이 초콜릿을 포장하는 공장이다.

#### 반복 작업을 통해 초콜릿을 포장하는데, 불량 제품이 들어올 경우 포장을 건너뛰고 계속해서 작업을 이어가야 한다.
#### "정상 초콜릿" 과 "불량 초콜릿" 이 섞여 들어오는 공정에서 정상 초콜릿은 포장을, 불량 초콜릿은 포장을 건너뛰는 코드는 다음과 같이 작성할 수 있다.

```python
import time

product = ['정상 초콜릿', '정상 초콜릿', '불량 초콜릿', '정상 초콜릿', '불량 초콜릿', '정상 초콜릿']

for i in product:
    if i == '불량 초콜릿':
        print("\n 불량 발견 포장 X")
        continue
    print('\n',i)
    print("포장 완료")
    time.sleep(1)
```

####  
### Continue 개념과 순서
#### continue 는 프로그램을 제어하기 위한 명령어로써, "건너 뛰다" 라는 의미를 갖고 있음.
#### 반복문 내에서 continue 를 만나게 되면 continue 아래에 있는 모든 반복문 내에 명령어들을 건너뜀

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

- continue는 break와 다르게 반복문을 종료하는 것이 아니라 그 시점에서의 반복문 순서를 건너뛰는 것임.

#### 
#### continue 를 사용하여 0부터 10까지 숫자 중 홀수를 출력하는 경우

```python
for i in range(0,11):
    if i % 2 == 0 :
        continue
    print(i)
```

## 3. pass 제어 명령어

#### pass 는 파이썬에서 아무것도 하지 않고 지나가기 위한 명령어임
#### 그럼 왜 필요할까?
1. 코드를 작성할 때, 지금은 당장 코드를 넣지 않을 것이지만 나중에 넣어야 하는 경우(오류가 나지 않게)
2. 특히, 나중에 배울 함수 or 클래스 에서 __init 과 같이 정의를 꼭 해야하지만 당장 하지 않을 때 사용함
#### 그래서 pass는 미완성이지만 코드를 오류 없이 돌아가게 하기 위한 용도로 쓰임

```python
for i in range(5):
    if i == 3:
        pass
    print(i)
```

#### 위와 같이 pass 는 i == 3인 경우에도 그냥 지나가서 3을 출력함
