## 2. 데이터형을 숫자로 바꿔주는 함수 : int(), float() 함수
- int() 함수는 데이터형을 정수인 integer로 바꿔주는 함수임    
- int() 함수 괄호 안에 정수형으로 변환할 데이터 혹은 변수를 넣어서 사용할 수 있음

```python
a = "3"
print(type(a))
```

- int() 함수를 사용하여 변환된 값을 다시 변수에 대입하여 출력할 수 있음

```python
a = int(a)
print(a)
print(type(a))
```

#### 
#### 문제 : 아래와 같은 문자열을 정수형으로 변환하여 데이터형과 같이 출력해보세요

```python
num = "12345"
print(num)
```

#### 
- int() 함수는 문자열뿐만 아니라 실수, 정수도 변환해줄 수 있음
#### 


```python
num1 = int("12345")
num2 = int(3.14)
num3 = int(1000)

print(num1)
print(num2)
print(num3)
```

#### * "12345" 와 같이 str 인 경우 int()를 사용하여 정상적으로 변환되었으나 3.14와 같은 float 데이터형인 경우에는 소수점이 사라지는 것을 볼 수 있음
#### * 또한 문자열이더라도 숫자가 아닌 경우에는 오류를 출력함

```python
name = int("고영민")
print(name)
```

#### 
- float() 함수는 int() 함수와 같이 데이터형을 변환하는 함수로 괄호 안에 받은 값을 실수인 float으로 변환해주는 함수임

```python
num2 = float(3.14)
print(num2)
print(type(num2))
```

```python
num2 = int(3.14)
print(type(num2))
print(num2)

num2 = float(3.14)
print(type(num2))
print(num2)
```

## 3. str() 과 bool() 함수
- int(), float() 함수와 같이 입력받은 값을 숫자 데이터형으로 변환해주는 함수처럼 str()과 bool() 함수는 각각 문자열과, 불 데이터형으로 변환해주는 함수임

str() 함수 사용 예시

```python
a = 3

a = str(a)
print(a)
print(type(a))
```

bool() 함수 사용 예시 -> bool 형은 0 : False , 1 : True 로 인식함

```python
a = 0

a = bool(a)
print(a)
print(type(a))
```

```python
a = 1

a = bool(a)
print(a)
print(type(a))
```

- bool 데이터 형은 0 또는 1 숫자로 인식하기 때문에 다시 int(), float() 을 사용하여 데이터형 변환을 수행할 수 있음.

```python
a = True

a = int(a)
print(a)
print(type(a))
```

```python
a = True

a = float(a)
print(a)
print(type(a))
```

## 4. bin() 함수
- bin() 함수는 정수형 데이터를 이진수로 변환하여 보여주는 함수임   
- bin() 함수를 사용하여 실제 컴퓨터의 메모리에 저장되는 값의 형태를 알 수 있음   
- 정수형 데이터인 경우에만 bin()을 사용 가능함

정수 2는 이진수로 '10' 이고 이진수임을 알리는 표시로 '0b' 를 붙여서 '0b10' 으로 표시됨   
또한 표시되는 데이터는 str 로 표시됨

```python
a = bin(2)
print(a)
print(type(a))
```

```python
a = 'a'
a = bin(a)
print(a)
```

- 위와 같이 문자열 'a' 를 bin()으로 변환할 수 없는 경우, ord() 함수를 사용하여 정수형으로 변환한 뒤, bin()을 통해 다시 변환할 수 있음   
- ord() 함수는 문자열을 유니코드로 반환하는 함수
- 유니코드(unicode) : 전 세계의 모든 문자를 다루도록 설계된 표준 문자 전산 처리 방식

```python
a = 'a'
a = ord(a)
print(a)
```

```python
a = bin(a)
print(a)
```

## 5. 사용자에게 문자열을 입력받고 print()을 사용하여 다른 문자열과 함께 출력하기
- input()을 사용하여 입력받은 문자열을 print() 함수로 다른 문자열들과 함께 출력할 경우 "," 쉼표로 구분하여 출력할 수 있음

```python
user_name = input("이름을 입력하세요 : ")
user_number = input("휴대전화 번호를 입력하세요 : ")

print("이름은 ", user_name, " 이고 번호는 ", user_number, "입니다.")
```

- 또한 같은 str 데이터형이므로 str끼리는 덧셈이 가능하므로 print() 함수 안에 + 를 사용하여 출력할 수 있음

```python
user_name = input("이름을 입력하세요 : ")
user_number = input("휴대전화 번호를 입력하세요 : ")

print("이름은 " + user_name + " 이고 번호는 " + user_number + "입니다.")
```
