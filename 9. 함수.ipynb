```python
def big(x,y,z):
    big_n = x
    
    if x > y :
        if x > z :
            big_n = x
        else :
            big_n = z
    else:
        if y > z :
            big_n = y
        else : 
            big_n = z
            
    return big_n
```

```python
a = big(x=10, y=15, z=5)
print(a)
```

```python
def cal(x,y):
    r1 = x + y
    r2 = x - y
    r3 = x * y
    r4 = x / y
    return r1, r2, r3, r4
```

```python
a,b,c,d = cal(x=10, y=5)
print(" : ",a)
print(" : ",b)
print(" : ",c)
print(" : ",d)
```

-  ,         

```python
def average_cal(*args):
    length = len(args)
    
    total = 0
    for i in args:
        total = total + i
    
    avg = total / length
    return avg
```

```python
avg = average_cal(10, 20, 30)
print(avg)
```

```python
avg = average_cal(10, 20, 30, 50 ,100)
print(avg)
```
